```python
def percentile(stock,stock1,days,x):
    maximum = []
    minimum = []
    for y in range(len(stock)) :
        if y  < len(stock)-(days-1) and y >=x-(days-1):
            p = (max(stock.iloc[y:y+days].values))
            q = (min(stock.iloc[y:y+days].values))
            maximum.append(p[0])
            minimum.append(q[0])
    
    stock1["Past " + str(days) + "max"]= maximum
    stock1["Past " + str(days) + "min"]= minimum
    ay = pd.DataFrame(maximum)
    by = pd.DataFrame(minimum)
    cy = (by + (ay-by)*0.75).values
    dy = (by + (ay-by)*0.25).values
    stock1["75th percentile for "+ str(days) + "days"] = cy
    stock1["25th percentile for "+ str(days) + "days"] = dy

percentile(BHARTIARTL60,BHARTIARTL1,60,60)
percentile(BHARTIARTL120,BHARTIARTL1,120,120)
percentile(BHARTIARTL180,BHARTIARTL1,180,180)
percentile(BHARTIARTL252,BHARTIARTL1,252,252)
BHARTIARTL1.head()
```
